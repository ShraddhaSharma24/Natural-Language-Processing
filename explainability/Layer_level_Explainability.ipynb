```python
from transformers import BertTokenizer, BertForMaskedLM
import torch

# Load pre-trained model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForMaskedLM.from_pretrained('bert-base-uncased')

# Move model to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
```

```python
#verb tense probing
sentences_tense = [
    "I walked to the store.",      # Past tense
    "I walk to the store.",       # Present tense
    "I will walk to the store."   # Future tense
]

# Tokenize sentences
inputs = tokenizer(sentences_tense, return_tensors='pt', padding=True, truncation=True).to(device)

# Pass through BERT model to get predictions
with torch.no_grad():
    outputs = model(**inputs)

# Print model outputs (logits for prediction analysis)
print(outputs.logits)
```

```python
#Gender pronoun probing
sentences_pronouns = [
    "The doctor said he would arrive soon.",  # He (male pronoun)
    "The doctor said she would arrive soon."  # She (female pronoun)
]

# Tokenize sentences
inputs = tokenizer(sentences_pronouns, return_tensors='pt', padding=True, truncation=True).to(device)

# Pass through BERT model
with torch.no_grad():
    outputs = model(**inputs)

# Print model outputs (logits for prediction analysis)
print(outputs.logits)
```

```python
#World Knowledge probing
sentences_world_knowledge = [
    "The capital of France is Paris.",    # Factual sentence
    "The capital of France is Berlin."   # Factual mistake
]

# Tokenize sentences
inputs = tokenizer(sentences_world_knowledge, return_tensors='pt', padding=True, truncation=True).to(device)

# Pass through BERT model
with torch.no_grad():
    outputs = model(**inputs)

# Print model outputs (logits for prediction analysis)
print(outputs.logits)
```

```python
import torch.nn.functional as F

# Example of extracting probabilities from logits for the first sentence in the world knowledge test
probabilities = F.softmax(outputs.logits, dim=-1)

# Print probabilities for a specific token (for example, checking 'Paris' vs 'Berlin' for world knowledge)
print(probabilities[0, 0, tokenizer.encode("Paris", add_special_tokens=False)[0]].item())
print(probabilities[0, 1, tokenizer.encode("Berlin", add_special_tokens=False)[0]].item())
```

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from captum.attr import LayerIntegratedGradients
```

```python
model_name= "bert-base-uncased"
tokenizer=BertTokenizer.from_pretrained(model_name)
model=BertForSequenceClassification.from_pretrained(model_name)
model.eval()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
sentence ="I absolutely loved the movie!"
inputs = tokenizer(sentence, return_tensors='pt', truncation=True).to(device)
```

```python
#helper fuunction for captum
def predict(inputs):
    outputs = model(**inputs)
    return torch.softmax(outputs.logits, dim=-1)

def forward_func(input_embeds):
    attention_mask = torch.ones(input_embeds.size()[:-1]).to(device)
    output = model(inputs_embeds=input_embeds, attention_mask=attention_mask)
    return torch.softmax(output.logits, dim=-1)

# Choose a target label (0 or 1) for explanation
target_label = 1  # Suppose 1 = Positive Sentiment
```

```python
# Initialize Layer Integrated Gradients
lig = LayerIntegratedGradients(forward_func, model.bert.embeddings)

# Get input embeddings
input_ids = inputs["input_ids"]
embeddings = model.bert.embeddings(input_ids)

# Compute Layer-wise attributions
attributions, delta = lig.attribute(inputs=embeddings,
                                    baselines=torch.zeros_like(embeddings),
                                    target=target_label,
                                    return_convergence_delta=True)

# Summing over the embedding dimension
layer_attributions = attributions.sum(dim=-1).squeeze(0).detach().cpu()

# Visualize
print(layer_attributions)
```

```python
def layer_forward_func(hidden_states):
    attention_mask = torch.ones(hidden_states.size()[:-1]).to(device)
    output = model(inputs_embeds=hidden_states, attention_mask=attention_mask)
    return torch.softmax(output.logits, dim=-1)

# Now, pick a layer
layer_module = model.bert.encoder.layer[8].output

# Layer Integrated Gradients for layer 8
lig = LayerIntegratedGradients(layer_forward_func, layer_module)

# Attribution for that layer
attributions, delta = lig.attribute(inputs=embeddings,
                                    baselines=torch.zeros_like(embeddings),
                                    target=target_label,
                                    return_convergence_delta=True)

# Sum across tokens and dimensions
layer_score = attributions.sum().item()
print(f"Layer 8 Attribution Score: {layer_score}")
```
